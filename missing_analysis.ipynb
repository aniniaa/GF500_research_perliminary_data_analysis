## Carbon Credit difference analysis 

```python
import pandas as pd

df_2024 = pd.read_csv('2024.csv', encoding='latin-1')
df_netzero = pd.read_csv('netzero.csv', encoding='latin-1')

# Filter companies where CC = 1
cc_companies = df_2024[df_2024['will use cabon credits'] == 1]

# Merge with netzero data
comparison = cc_companies[['Company Name', 'will use cabon credits']].merge(
    df_netzero[['company_name', 'Carbon_credits']], 
    left_on='Company Name',
    right_on='company_name',
    how='left'
)

comparison.to_excel('cc_netzero_comparison.xlsx', index=False)
print(comparison['Carbon_credits'].value_counts())
```
